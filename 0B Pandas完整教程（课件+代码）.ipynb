## 为什么学习 Pandas？
- NumPy 能够帮我们处理**数值型数据**，但这还不够。很多时候，我们的数据除了数值之外，还有**字符串**等。
- Pandas 是 Python 语言的一个扩展程序库，用于**数据分析**。
- Pandas 名字衍生自术语 **panel data**（面板数据）和 **Python data analysis**（Python 数据分析）。

## Pandas 的特点
- 强大的分析**结构化数据**的工具集，基础是 NumPy（提供高性能的矩阵运算）。
- 可以从各种文件格式（CSV、JSON、SQL、Microsoft Excel）**导入数据**。
- 可以对各种数据进行运算操作：归并、再成形、选择、数据清洗和数据加工特征。
- 广泛应用在**学术、金融、统计学**等各个数据分析领域。


## 数据类型介绍

### 横截面数据（Cross-sectional Data）
**某一时间点**收集的**不同对象**的数据。

### 时间序列数据（Time Series Data）
**同一对象**在**不同时间**连续观察所得的数据。

### 面板数据（Panel Data）
前两者合起来的数据。既可以分析个体之间的差异情况，又可以描述个体的动态变化特征。


## Pandas 中最重要的两个元素

| | NumPy | Pandas |
|------|-------|--------|
| 核心元素 | 数组（array） | **Series**（一维）、**DataFrame**（二维） |

- **Series**：类似于一维数组的对象，由一组数据 + 一组与之相关的数据标签（索引）组成。索引在左，数据在右。
- **DataFrame**：表格型数据结构，含有一组有序的列，每列可以是不同值类型。既有**行索引**也有**列索引**，可以看做由 Series 组成的字典。

## 一、Series

Series 是一种**一维标记的数组型对象**，能够保存任何数据类型（整数、字符串、浮点数、Python 对象等），包含了数据标签，即索引。

**组成：索引在左，数据在右。索引是自动创建的。**

### 1.1 创建 Series

````python
# 安装 pandas（命令行执行）
# pip install pandas
import pandas as pd
import numpy as np
````

````python
# 方法1：通过列表创建
s1 = pd.Series([1, 2, 3, 4, 5])
s1  # 左边一列为自动生成的索引（0,1,2...）
````

````python
# 查看 Series 的 values 和类型
print('values:', s1.values)
print('type:', type(s1))

# 对比 NumPy 数组
a = np.array([1, 2, 3, 4, 5])
print('numpy array type:', type(a))
````

````python
# 方法2：通过数组创建
pd.Series(a)
````

````python
# 手动指定索引
s2 = pd.Series(a, index=['a', 'b', 'c', 'd', 'e'])
s2

# 索引个数必须与数据个数一致！
````

````python
# 分别获取索引和数据
print('index:', s2.index)
print('values:', s2.values)
print('index type:', type(s2.index))   # pandas.core.indexes.base.Index
print('values type:', type(s2.values)) # numpy.ndarray（Pandas 构建在 NumPy 之上）
````

````python
# 方法3：通过字典创建（key → index, value → data）
d = {'name': 'Mike', 'age': 18, 'score': 80}
s3 = pd.Series(d)
s3
````

````python
# 通过字典创建时可以指定 index（多出的 key 对应 NaN）
s3 = pd.Series(d, index=['name', 'age', 'score', 'id'])
s3
````

### 1.2 检查缺失值

````python
# isnull / notnull
print('isnull:', s3.isnull().tolist())
print('notnull:', s3.notnull().tolist())
````

### 1.3 通过索引获取数据

````python
# 通过标签名（类似字典）
print(s3['score'])
print(s3.get('sc', 1))  # get 方法：不存在时返回默认值
````

````python
# 通过位置下标（推荐使用 iloc）
print('iloc[1]:', s3.iloc[1])

# 取多个值
print(s3[['age', 'score']])     # 传入标签列表
print(s3.iloc[[1, 3]])         # 传入位置列表
````

````python
# 切片（下标切片不包含最后一项，标签切片包含最后一项）
print('标签切片 [name:score]:\n', s3['name':'score'])
print('位置切片 [1:3]:\n', s3.iloc[1:3])
````

### 1.4 按条件查找数据

````python
# 创建带标签的 Series
s = pd.Series(range(5), index=list('abcde'))
print('s:', s.tolist())

# 布尔索引
print('s < 3:\n', s[s < 3])
````

````python
# where 函数（Series 的方法，与 numpy.where 不同）
print('where(s>3):')
print(s.where(s > 3, 0))  # 符合条件的保留，否则变为指定值

# mask 函数：与 where 相反
print('mask(s>3):')
print(s.mask(s > 3))  # 符合条件的变为 NaN
````

### 1.5 索引与数据的对应关系不受运算影响

````python
# 运算结果保持索引-数据的对应关系
print('s2:', s2.tolist())
print('s2 * 2:', (s2 * 2).tolist())
````

### 1.6 name 属性

````python
s2.name = 'temp'            # Series 的名字
s2.index.name = 'year'      # index 列的名字
s2
````

### 1.7 head() 和 tail() 方法

````python
print('head(3):\n', s2.head(3))
print('tail(2):\n', s2.tail(2))
````

### 1.8 增删改

````python
# 增
s = pd.Series([1, 2, 3, 4, 5])
s['e'] = 6  # 直接赋值新增
print('新增后:', s.tolist())

# append（返回新 Series，不修改原对象）
s2 = s.append(pd.Series([7], index=['f']))
print('append 后:', s2.tolist())
````

````python
# 删：del 和 drop
del s[4]  # 删除 index=4 的元素
print('del 后:', s.tolist())

# drop：inplace=True 直接修改，inplace=False 返回新对象
s.drop(2, inplace=True)
print('drop 后:', s.tolist())
````

````python
# 改
s[0] = 100
print('修改后:', s.tolist())
````

## 二、DataFrame

DataFrame 是一个**表格型**的数据结构，含有一组有序的列。
- 每列可以是不同的值类型（数值、字符串、布尔型值）。
- DataFrame 既有**行索引**也有**列索引**。
- 可以被看做由 Series 组成的字典（共同用一个索引）。

### 2.1 创建 DataFrame

**常见创建方式：**
- 从数据源直接读取（CSV、Excel 等）
- **通过字典**：以列表/数组/元组为元素的字典、以 Series 为元素的字典、以字典为元素的字典
- **通过列表**：二维列表/数组、字典列表、Series 列表

````python
# 从 CSV 读取
df = pd.read_csv('ex1.csv')
df
````

````python
# 查看 DataFrame 的基本属性
print('index:', df.index.tolist())
print('columns:', df.columns.tolist())
print('values:\n', df.values)
````

#### 2.1.1 通过以列表/数组/元组为元素的字典（长度必须相同）

````python
data = {
    'a': [1.0, 2, 3, 4],
    'b': (5, 6, 7, 8),
    'c': np.arange(9, 13)
}
df = pd.DataFrame(data)
df
````

````python
# 创建时指定 index 和 columns
df = pd.DataFrame(data, index=['A', 'B', 'C', 'D'], columns=['a', 'b', 'c', 'd'])
df  # d 列全为 NaN（字典中没有 d 这个 key）
````

#### 2.1.2 通过以 Series 为元素的字典

````python
# Series 长度可以不同，缺失位置为 NaN
df = pd.DataFrame({
    'a': pd.Series([1, 2, 3]),
    'b': pd.Series(np.arange(3, 5)),
})
df
````

````python
# 不同 index 的 Series 会堆叠到一起
df = pd.DataFrame({
    'a': pd.Series([1, 2, 3], index=['e', 'f', 'g']),
    'b': pd.Series(np.arange(3, 5)),
})
df
````

#### 2.1.3 通过以字典为元素的字典（最外层 key → columns，内层 key → index）

````python
d = {
    'a': {'apple': 3.6, 'banana': 5.6},
    'b': {'apple': 9.6, 'banana': 3.2},
    'c': {'apple': 7.7}
}
df = pd.DataFrame(d)
df
````

#### 2.1.4 通过二维数组

````python
a = np.arange(12).reshape(3, 4)
df = pd.DataFrame(a, index=list('abc'), columns=list('ABCD'))
df
````

#### 2.1.5 通过字典构成的列表（内层 key → columns）

````python
a = [
    {'apple': 1, 'banana': 2},
    {'apple': 3, 'banana': 4},
    {'apple': 5, 'banana': 6}
]
pd.DataFrame(a)
````

#### 2.1.6 通过 Series 构成的列表

````python
l2 = [
    pd.Series(np.arange(3)),
    pd.Series([4, 5, 6])
]
pd.DataFrame(l2, index=['a', 'b'], columns=[0, 1, 2])
````

### 2.2 DataFrame 基本用法

````python
df = pd.DataFrame(np.arange(9).reshape(3, 3), index=['a', 'b', 'c'], columns=['A', 'B', 'C'])
df
````

````python
# 转置
df.T
````

````python
# 通过列索引获取列数据（返回 Series）
print(df['A'])
print('type:', type(df['A']))  # pandas.core.series.Series

# 取多列（返回 DataFrame）
print(df[['A', 'B']])
print('type:', type(df[['A', 'B']]))  # pandas.core.frame.DataFrame
````

### 2.3 高级索引：loc 和 iloc

- **loc**：标签索引（按 index/columns 的标签值）
- **iloc**：位置索引（按整数位置）

规则和 NumPy 二维数组相同：最外层逗号，左边是行，右边是列。

````python
df = pd.DataFrame(np.random.randint(0, 10, size=(3, 4)), 
                  index=list('abc'), columns=list('ABCD'))
df
````

````python
# loc：标签索引
print('单个值 loc[a,A]:', df.loc['a', 'A'])
print('整行 loc[a]:\n', df.loc['a'].tolist())
print('整列 loc[:,A]:\n', df.loc[:, 'A'].tolist())
print('多行 loc[[a,c]]:\n', df.loc[['a', 'c']])
print('数据块 loc[a:c, [B,D]]:\n', df.loc['a':'c', ['B', 'D']])
````

````python
# iloc：位置索引
print('整行 iloc[0]:\n', df.iloc[0].tolist())
print('整列 iloc[:,0]:\n', df.iloc[:, 0].tolist())
print('数据块 iloc[0:2, 1:3]:\n', df.iloc[0:2, 1:3])
````

````python
# 查：where 和 mask
df = pd.DataFrame(np.arange(20).reshape(4, 5), 
                  columns=['E', 'A', 'B', 'C', 'D'])
print('where(df>5, -1):\n', df.where(df > 5, -1))
print('mask(df>5):\n', df.mask(df > 5))
````

### 2.5 字符串操作

#### Python 内置字符串方法

````python
# split：分割
val = 'a,b, c'
print(val.split(','))

# strip：去除首尾空白 + 列表推导式
print([x.strip() for x in val.split(',')])

# join：连接
a, b, c = 'how', 'are', 'you'
print(' '.join([a, b, c]))

# 判断子串
s = 'how are you'
print('are' in s)       # True
print(s.find('aree'))    # -1（不存在）

# replace、title
print(s.replace('are', 'ARE'))
print(s.title())
````

#### Pandas 中的字符串处理函数（.str 访问器）

````python
df = pd.DataFrame([list('abc'), list('def'), list('egh')])
df
````

````python
# str.cat：拼接
print(df[0].str.cat())           # 'ade'
print(df[0].str.cat(sep=','))   # 'a,d,e'

# str.contains：判断是否包含
print(df[0].str.contains('a'))

# 字符串条件过滤（配合布尔索引）
con = df['A'].str.contains('bc')
df[con]
````

## 三、Pandas 的索引（Index）

### 3.1 Index 对象
- Series 和 DataFrame 中的索引都是 **Index 对象**。
- **不可直接修改** Index 对象的元素（保证了数据安全）。

### 常见的 Index 种类
1. **Index**：普通索引
2. **Int64Index**：整数索引
3. **MultiIndex**：层级索引
4. **DatetimeIndex**：时间戳索引

````python
# Index 对象不可修改
ps = pd.Series(range(5), index=['a', 'b', 'c', 'd', 'e'])
print('index type:', type(ps.index))
print('index:', ps.index.tolist())
# ps.index[0] = 'f'  # TypeError: Index does not support mutable operations
````

````python
# 整数索引
ps = pd.Series(range(5), index=[0, 1, 2, 3, 4])
print('Int64Index type:', type(ps.index))
````

### 3.2 reindex（重索引）
对索引进行重新排序，自动对应数据。

````python
ps = pd.Series([1, 2, 3, 4], index=['b', 'c', 'a', 'd'])
print('原始:', ps.tolist())

# 对索引排序
ps2 = ps.reindex(['a', 'b', 'c', 'd'])
print('reindex 后:', ps2.tolist())
````

````python
# DataFrame 的 reindex
df = pd.DataFrame(np.arange(9).reshape(3, 3), index=['b', 'c', 'a'], columns=['A', 'B', 'C'])
print(df)

# 行 reindex
df2 = df.reindex(['a', 'b', 'c', 'd', 'e'])
print('行 reindex:\n', df2)

# 列 reindex
df3 = df.reindex(columns=['C', 'B', 'A'])
print('列 reindex:\n', df3)
````

````python
# fill_value：填充缺失值
print(df.reindex(['a', 'b', 'c', 'd'], fill_value=0))

# method：ffill（向前填充）/ bfill（向后填充）
df1 = pd.DataFrame(np.arange(9).reshape(3, 3), index=['a', 'b', 'e'])
print('ffill:\n', df1.reindex(['a', 'b', 'c', 'e'], method='ffill'))

# 注意：method 要求原 index 单调递增或递减！
````

### 3.3 DatetimeIndex（时间索引）

````python
# 创建时间索引
date_index = pd.date_range('1/1/2010', periods=6, freq='D')
print('时间索引:', date_index)
print('type:', type(date_index))

# 构造以时间为 index 的 DataFrame
df2 = pd.DataFrame({'prices': [100, 101, np.nan, 100, 89, 88]}, index=date_index)
df2
````

````python
# 扩展时间范围
date_index2 = pd.date_range('12/29/2009', periods=10, freq='D')
df2.reindex(date_index2, method='bfill')
````

### 3.4 修改索引

````python
df = pd.DataFrame(np.arange(12).reshape(3, 4), index=['a', 'b', 'c'], 
                  columns=['one', 'two', 'three', 'four'])
df
````

````python
# 方法1：整体赋值
df.index = ['Beijing', 'Shandong', 'Hebei']
df.columns = ['aa', 'bb', 'cc', 'dd']
df
````

````python
# 方法2：使用 map 函数
df.index = df.index.map(lambda x: x[:4].upper())
df
````

````python
# 方法3：使用 rename
df.rename(index={'BEIJ': 'BJ', 'HEBE': 'HB', 'SHAN': 'SD'}, inplace=True)
df

# rename 也可以传入函数
df.rename(index=lambda x: x.lower(), inplace=True)
df
````

````python
# 方法4：set_index / reset_index（将已有列设为行索引）
df = pd.DataFrame(np.arange(9).reshape(3, 3), index=['a', 'b', 'c'], columns=['A', 'B', 'C'])
print('原始:\n', df)

# 将 A 列设为行索引
df1 = df.set_index(['A'])
print('set_index(A):\n', df1)

# 双层索引
df1 = df.set_index(['A', 'B'])
print('双层索引:\n', df1)

# 重置索引
df1.reset_index()
````

## 四、对齐运算

Pandas 中两个 Series/DataFrame 做运算时，会**自动对齐索引**。如果某个索引只在一方存在，结果中该位置为 NaN。

````python
# Series 对齐
s1 = pd.Series(np.arange(4), index=list('abcd'))
s2 = pd.Series(np.arange(5), index=list('aecbd'))
print('s1:', s1.tolist())
print('s2:', s2.tolist())
print('s1 + s2:\n', (s1 + s2).tolist())
````

````python
# DataFrame 对齐
df1 = pd.DataFrame(np.arange(12).reshape(3, 4), index=list('abc'), columns=list('ABCD'))
df2 = pd.DataFrame(np.arange(9).reshape(3, 3), index=list('acf'), columns=list('ABD'))
print('df1:\n', df1)
print('df2:\n', df2)
print('df1 + df2:\n', df1 + df2)
````

### 4.1 使用填充值的算术方法

为避免 NaN，可以使用 `add`、`sub` 等方法的 `fill_value` 参数。

| 方法 | 描述 | 带 r 的反向方法 |
|------|------|----------------|
| add | 加法 (+) | radd |
| sub | 减法 (-) | rsub |
| div | 除法 (/) | rdiv |
| floordiv | 整除 (//) | rfloordiv |
| mul | 乘法 (*) | rmul |
| pow | 幂次方 (**) | rpow |

````python
# 带填充值的加法（将对方不存在的值视为 0）
print('s1.add(s2, fill_value=0):', s1.add(s2, fill_value=0).tolist())
print('df1.add(df2, fill_value=0):\n', df1.add(df2, fill_value=0))
````

### 4.2 Series 和 DataFrame 的混合运算

类似于 NumPy 的广播机制：后缘维度的轴长为 1 或者相等。

- **取行（Series）与 DataFrame 运算**：默认沿行方向广播（axis=0）
- **取列（Series）与 DataFrame 运算**：需要用 `sub` 等方法指定 `axis=0`

````python
df1 = pd.DataFrame(np.arange(12).reshape(3, 4), index=list('abc'), columns=list('ABCD'))
print('df1:\n', df1)

# 取一行（Series），默认按行广播
s_row = df1.iloc[0]
print('df1 - s_row:\n', df1 - s_row)  # 每行减去第0行
````

````python
# 取一列（Series），需要用 sub 指定 axis=0
s_col = df1.iloc[:, 0]
print('s_col:', s_col.tolist())
# df1 - s_col 会按列对齐（找不到匹配），需要指定 axis=0 按行广播
print('df1.sub(s_col, axis=0):\n', df1.sub(s_col, axis=0))
````

## 五、Pandas 中的常用函数

本节涵盖：apply/applymap、排序、唯一值和成员属性、去除重复数据、处理缺失数据、map 数据转换。

### 5.1 apply 和 applymap

- **apply**：将函数应用到**行或列**（axis=0 按列，axis=1 按行）
- **applymap / map**：将函数应用到**每个元素**

````python
df = pd.DataFrame(np.random.normal(size=(5, 4)))
df
````

````python
# apply：对每行操作
df.apply(lambda x: x.max() - x.min(), axis=1)  # 每行最大值减最小值
````

````python
# apply：对每列操作
df.apply(lambda x: x.mean(), axis=0)  # 每列平均值
````

````python
# applymap：对每个元素操作（新版推荐使用 map）
df.map(lambda x: round(x, 2))
````

### 5.2 排序

- **sort_index**：按索引排序
- **sort_values**：按数值排序（by 指定列，ascending 指定升/降序）

````python
# 索引排序
s1 = pd.Series(np.arange(4), index=list('dbca'))
print('原始:', s1.tolist())
print('sort_index:', s1.sort_index().tolist())

df = pd.DataFrame(np.arange(12).reshape(3, 4), index=list('bca'), columns=list('ADBC'))
print('\n原始:\n', df)
print('行排序 sort_index():\n', df.sort_index())
print('列排序 sort_index(axis=1):\n', df.sort_index(axis=1))
````

````python
# 按值排序
print('Series sort_values:', s1.sort_values().tolist())

# DataFrame 按值排序
df = pd.DataFrame(np.random.randint(0, 10, (3, 4)), 
                  index=list('abc'), columns=list('ABCD'))
print(df)
print('按 A 列排序:\n', df.sort_values(by='A'))
print('按多列排序:\n', df.sort_values(by=['A', 'B'], ascending=[True, False]))
````

````python
# 缺失值排序：默认在末尾，可用 na_position 控制
s1 = pd.Series([2, np.nan, 1, np.nan, 3])
print('NaN 在最前:\n', s1.sort_values(na_position='first').tolist())
````

### 5.3 唯一值和成员属性

### 5.5 处理缺失数据

### 5.6 利用字典或函数进行数据转换（map）

````python
data = pd.DataFrame({'food': ['Apple', 'banana', 'orange', 'apple', 'Mango', 'tomato'],
                    'price': [4, 3, 3.5, 6, 12, 3]})
data
````

````python
# 方法1：建立映射字典 + map
cat = {
    'apple': 'fruit',
    'banana': 'fruit',
    'orange': 'fruit',
    'Apple': 'fruit',
    'Mango': 'fruit',
    'tomato': 'vegetable'
}
data['food'] = data['food'].str.lower()  # 先统一大小写
data['category'] = data['food'].map(cat)
data
````

````python
# 方法2：使用函数映射
def trans(x):
    return 'high' if x > 4 else 'low'

data['level'] = data['price'].map(trans)
data
````

````python
# 使用 lambda 函数
data['level'] = data['price'].map(
    lambda x: 'low' if x < 4 else ('medium' if x < 10 else 'high')
)
data
````

## 七、数据连接/合并

### 7.1 pd.merge
根据单个或多个键将不同 DataFrame 的行连接起来。

```python
pd.merge(left, right, how='inner', on=None, left_on=None, right_on=None)
```

| 参数 | 说明 |
|------|------|
| left | 左边的 DataFrame |
| right | 右边的 DataFrame |
| how | 合并方式：inner / outer / left / right（默认 inner） |
| on | 两边共有的列名，作为连接键 |
| left_on | 左侧 DataFrame 中用作连接键的列 |
| right_on | 右侧 DataFrame 中用作连接键的列 |


````python
# 构造示例数据
df1 = pd.DataFrame({'key': ['A', 'B', 'C'], 'value1': [1, 2, 3]})
df2 = pd.DataFrame({'key': ['B', 'C', 'D'], 'value2': [4, 5, 6]})
print('df1:\n', df1)
print('df2:\n', df2)
````

````python
# 内连接（inner）：只保留两边都有的 key
print('inner join:\n', pd.merge(df1, df2, on='key', how='inner'))

# 左连接（left）：保留左边所有的 key
print('left join:\n', pd.merge(df1, df2, on='key', how='left'))

# 右连接（right）：保留右边所有的 key
print('right join:\n', pd.merge(df1, df2, on='key', how='right'))

# 外连接（outer）：保留两边的所有 key
print('outer join:\n', pd.merge(df1, df2, on='key', how='outer'))
````

### 7.2 pd.concat

沿轴方向拼接数据。
- `axis=0`：纵向拼接（上下堆叠）
- `axis=1`：横向拼接（左右拼接）
- `join='outer'`：取并集（默认）
- `join='inner'`：取交集

````python
# concat 示例
df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]}, index=['a', 'b'])
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]}, index=['c', 'd'])
print('纵向拼接 axis=0:\n', pd.concat([df1, df2]))

df3 = pd.DataFrame({'C': [9, 10], 'D': [11, 12]}, index=['a', 'b'])
print('横向拼接 axis=1:\n', pd.concat([df1, df3], axis=1))
````

## 八、数据的分组与聚合（groupby）

`df.groupby(分组列).agg(聚合函数)` 是 Pandas 中最常用的数据分析模式。

常用聚合函数：`mean`、`sum`、`count`、`max`、`min`、`std`、`var` 等。
也可通过 `agg()` 传入自定义函数。

````python
# 读取示例数据
df = pd.read_csv('fruit.csv')
df
````

````python
# 用循环理解分组逻辑
s = df['fruit']
s2 = df['price']
for i in s.unique():
    print(i, 'average price:', s2[s == i].mean())
````

````python
# groupby 基本用法
avg_price = df.groupby('fruit', as_index=False)['price'].mean()
avg_price
````

````python
# 使用 agg 自定义聚合函数
def diff(arr):
    return arr.max() - arr.min()

price_diff = df.groupby('fruit', as_index=False)['price'].agg(价格极差=diff)
price_diff
````

````python
# 同时使用多种聚合
df.groupby('fruit', as_index=False)['price'].agg(['mean', 'max', 'min', 'std'])
````

## 九、正则表达式速查表

Pandas 的 `.str` 访问器支持正则表达式，常用在 `contains`、`replace`、`extract` 等方法中。

### 字符匹配

| 实例 | 描述 |
|------|------|
| `[Pp]ython` | 匹配 "Python" 或 "python" |
| `[0-9]` | 匹配任何数字 |
| `[a-z]` | 匹配任何小写字母 |
| `[A-Z]` | 匹配任何大写字母 |
| `[a-zA-Z0-9]` | 匹配任何字母及数字 |
| `[^au]` | 除了 a 和 u 以外的所有字符 |
| `[^0-9]` | 匹配除了数字外的字符 |

### 元字符

| 实例 | 描述 |
|------|------|
| `.` | 匹配除 `\n` 之外的任何单个字符 |
| `?` | 匹配一个字符零次或一次 |
| `+` | 匹配 1 次或多次 |
| `*` | 匹配 0 次或多次 |
| `\d` | 匹配一个数字字符，等价于 `[0-9]` |
| `\D` | 匹配一个非数字字符 |
| `\s` | 匹配任何空白字符（空格、制表符、换页符等） |
| `\S` | 匹配任何非空白字符 |
| `\w` | 匹配包括下划线的任何单词字符，等价于 `[A-Za-z0-9_]` |
| `\W` | 匹配任何非单词字符 |
| `\b` | 匹配单词边界 |


````python
# 正则表达式在 Pandas 中的使用示例
s = pd.Series(['apple', 'Apple123', 'Banana', '123', 'cat', 'dog_1'])

# 匹配包含数字的
print('包含数字:')
print(s[s.str.contains(r'\d')])

# 匹配以 a 开头的（不区分大小写）
print('\n以 a 开头:')
print(s[s.str.contains(r'^a', case=False)])

# 提取数字
print('\n提取数字:')
print(s.str.extract(r'(\d+)'))
````
